```python
# Método Concat permite unir dataframes a lo largo de un eje que puede ser vertical u horizontal
import pandas as pd
```

```python
df1 = pd.DataFrame({'Nombre': ["Juan", "Gabriela", "Elena"], 'Edad': [23, 31, 21]})
df1
```

```python
df2 = pd.DataFrame({'Nombre': ["Carmela", "Max", "Laura"], 'Edad': [34, 25, 29]})
df2
```

```python
df_cancatenado = pd.concat([df1, df2])
df_cancatenado
```

```python
df_cancatenado2 = pd.concat([df1, df2], axis=1)
df_cancatenado2
```

```python
df_cancatenado = pd.concat([df1, df2], ignore_index=True)
df_cancatenado
```

```python
df_cancatenado = pd.concat([df1, df2], keys=['df1', 'df2'])
df_cancatenado
```

```python
df_cancatenado = pd.concat([df1, df2], keys=['enero', 'febrero'])
df_cancatenado
```

```python
#Concatenar DataFrames en Pandas con concat 1
ventas_enero = pd.DataFrame({'Producto': ["Manzanas", "Bananas", "Naranjas"], 'Cantidad': [300, 200, 150]})

ventas_febrero = pd.DataFrame({'Producto': ["Manzanas", "Bananas", "Naranjas"], 'Cantidad': [350, 210, 170]})
```

```python
#Tu tarea es utilizar el método concat() para concatenar estos dos DataFrames verticalmente en un nuevo DataFrame llamado ventas_total. 
#Asegúrate de ignorar los índices originales para que el 
#índice del DataFrame resultante sea continuo.

ventas_total = pd.concat([ventas_enero, ventas_febrero], ignore_index=True)
ventas_total
```

```python
#Concatenar DataFrames en Pandas con concat 2
datos_cliente = pd.DataFrame({'Nombre': ["Ana", "Luis", "Marta"], 'Edad': [34, 45, 28]})
compras_cliente = pd.DataFrame({'Producto': ["Libro", "Lápiz", "Cuaderno"], 'Precio': [15.50, 0.50, 2.00]})
```
