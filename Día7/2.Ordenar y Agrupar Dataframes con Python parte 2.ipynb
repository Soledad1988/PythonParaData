```python
import pandas as pd
```

```python
ruta = "C:/Users/brent/OneDrive/Escritorio/Python para Data Science & Machine Learning/Día7/Top-Películas.csv"
df = pd.read_csv(ruta)
```

```python
df
```

```python
# Ordenar DataFrames (por defectoo de menor a mayor)
df_ordenado = df.sort_values(by='rating')
df_ordenado.head(10)
```

```python
# Ordenar DataFrames de Mayor a menor
df_ordenado = df.sort_values(by='rating', ascending=False)
df_ordenado.head(10)
```

```python
# Ordenar mas de una columna, por ejemplo una lista
df_ordenado = df.sort_values(by=['rating', 'recaudación(M)'], ascending=False)
df_ordenado.head(10)
```

```python
# Agrupar Dataframes, agrupar es reunir datos basandose en categroias o valores comunes, útil para el análisis comparativo
df_agrupado = df.groupby('género')['rating'].mean()
df_agrupado
```

```python
# Total recaudado por año
df_agrupado = df.groupby('año')['recaudación(M)'].sum()
```

```python
df_agrupado.sort_values(ascending=False).head(10)
```
